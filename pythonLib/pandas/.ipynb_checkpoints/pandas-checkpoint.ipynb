### 2.数据预处理

```python
#加减乘除
print(info['age'])
div_10 = info['age']/10
print(div_10)
add_10 = info['age']+10
print(add_10)
print(info['age'].max())
print("------------")

#排序,是否新生成一个
info.sort_values('age',inplace=True)
print(info)
print("------------")
#降序
info.sort_values('age',inplace=True,ascending=False)
print(info)
```

```python
train = pd.read_csv("train.csv")
train.head()
age = train["Age"]
#缺失值处理
age_is_null = pd.isnull(age)
age_null_true = age[age_is_null]
print(len(age_null_true))

mean_age = sum(age)/len(age)
print(mean_age)

good_ages = train['Age'][age_is_null == False]
correct_mean_age = sum(good_ages)/len(good_ages)
print(correct_mean_age)
#自动过滤掉nan
print(train['Age'].mean())
```

```python
#计算三个不同等级舱的平均价格
passenger_classes = [1, 2, 3]
fares_by_class={}
for this_class in passenger_classes:
    pclass_rows = train[train["Pclass"]==this_class]
    pclass_fares = pclass_rows['Fare']
    fare_for_class = pclass_fares.mean()
    fares_by_class[this_class] = fare_for_class
print (fares_by_class)
#数据透视表
passenger_survial=train.pivot_table(index='Pclass',values='Survived',aggfunc=np.mean)
print(passenger_survial)
#aggfunc默认是均值
passenger_age=train.pivot_table(index='Pclass',values='Age')
print(passenger_age)
#不同登船地点统计
port_stats=train.pivot_table(index='Embarked',values=['Fare','Survived'],aggfunc=np.sum)
print(port_stats)

#去掉nan值
drop_na_columns = train.dropna(axis=1)
#如果年龄和性别有nan，则去掉
new_train_survival=train.dropna(axis=0,subset=['Age','Sex'])
print (new_train_survival)
#定位第几号记录
print(train.loc[83,'Age'])
print(train.loc[777,'Pclass'])
#安招年龄排序
new_titanic_survival=train.sort_values('Age',ascending=False)
print(new_titanic_survival.head(10))

#重置索引
titanic_reindexed = new_titanic_survival.reset_index(drop=True)
print(titanic_reindexed.head(10))

```

```python
#自定义函数
def hundredth_row(column):
    hundredth_item = column.loc[99]
    return hundredth_item
hundredth_row = train.apply(hundredth_row)
print(hundredth_row)
#same as
train.loc[99]

def is_minor(row):
    if row['Age']<18:
        return True
    else:
        return False

minors = train.apply(is_minor,axis=1)
print(minors)
```

### 3. series

```python
#相当于numpy的ndarry,包括一个index和一个value
```

```python
fandango = pd.read_csv('fandango_score_comparison.csv')
series_film = fandango['FILM']
print(type(series_film))
print(fandango.head())
print(series_film[0:5])
print('_______________')
series_rt = fandango['RottenTomatoes']
print(series_rt[0:5])
print('_______________')
from pandas import Series
film_names = series_film.values
print(film_names)
print(type(film_names))
print('_______________')

rt_scores=series_rt.values
series_custom = Series(rt_scores,index=film_names)
print(series_custom[['Top Five (2014)','Ant-Man (2015)']])

print('_______________')
fiveten = series_custom[5:10]
print(fiveten)

print('______fileNamelist_________')
original_index = series_custom.index.tolist()
print(original_index)
print('_______________')
sorted_index = sorted(original_index)
sorted_by_index=series_custom.reindex(sorted_index)
print(sorted_by_index)
```

```python
#作用同上，根据index或者values排序
sc2 = series_custom.sort_index()
print(sc2.head())
print('----------------------------------------')
sc3 = series_custom.sort_values()
print(sc3[0:10])
```

```python
#series结构传入到numpy结构中做计算,貌似是values做计算，index不参与计算的，注意下面的数字是values
print(np.add(series_custom,series_custom))
print(np.sin(series_custom))
print(np.max(series_custom))
```
